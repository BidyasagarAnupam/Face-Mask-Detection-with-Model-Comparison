# **Import libraries**

```python
import os
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, BatchNormalization, MaxPooling2D, Dropout
from tensorflow.keras.applications import ResNet50,Xception,EfficientNetB0,VGG16
```

# **load dataset**

```python
path = 'Face Mask Dataset'
batch_size = 16
image_size = (128, 128)
```

# Spliting data into test and train

```python
# Data Generators (without additional augmentation)
train_datagen = ImageDataGenerator(rescale=1.0 / 255)
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

train_generator = train_datagen.flow_from_directory(
    os.path.join(path, 'Train'),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = test_datagen.flow_from_directory(
    os.path.join(path, 'Validation'),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    os.path.join(path, 'Test'),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

# Transfer learning on pre-trained model

```python

def build_and_train_model(base_model, model_name, epochs=10, save_dir='saved_models'):
    # Freeze base model layers
    for layer in base_model.layers:
        layer.trainable = False

    # Build the model
    model = Sequential([
        base_model,
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(2, activation='softmax')  # Softmax for binary classification
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        validation_data=val_generator,
        steps_per_epoch=len(train_generator),
        validation_steps=len(val_generator),
        epochs=epochs
    )
    print(f"\n{model_name} Training Complete\n")

    # Ensure the save directory exists
    os.makedirs(save_dir, exist_ok=True)
    # Save the trained model
    model_path = os.path.join(save_dir, f"{model_name}.h5")
    model.save(model_path)
    print(f"Model saved at {model_path}")
    
    return model, history
```

# Load Pre-trained models and freezing the initial layers

```python
models = {
    "ResNet50": ResNet50(include_top=False, input_shape=(224, 224, 3)),
    "Xception": Xception(include_top=False, input_shape=(299, 299, 3)),
    "EfficientNetB0": EfficientNetB0(include_top=False, input_shape=(224, 224, 3)),
    "VGG16": VGG16(include_top=False, input_shape=(224, 224, 3))
}
```

# Plot a graph of different models with their accuracy and loss

```python

# Training models and collecting histories
histories = {}
trained_models = {}

for name, base_model in models.items():
    print(f"Training model: {name}")
    trained_model, history = build_and_train_model(base_model, name)
    histories[name] = history
    trained_models[name] = trained_model

# Plot training/validation accuracy and loss
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for idx, metric in enumerate(['accuracy', 'loss']):
    for i, (name, history) in enumerate(histories.items()):
        ax = axes[idx, i % 2]
        ax.plot(history.history[metric], label=f'{name} Train')
        ax.plot(history.history[f'val_{metric}'], label=f'{name} Validation')
        ax.set_title(f'{metric.capitalize()} - {name}')
        ax.set_xlabel('Epochs')
        ax.set_ylabel(metric.capitalize())
        ax.legend()

plt.suptitle("Model Training & Validation Metrics", size=18)
plt.show()
```

# Comparing the test accuracy of each models

```python
# Evaluate models on the test set
test_accuracies = {}
for name, model in trained_models.items():
    test_loss, test_acc = model.evaluate(test_generator)
    test_accuracies[name] = test_acc
    print(f"{name}: Test Accuracy = {test_acc:.2f}")
```

# plotting test accuracy of each model

```python
# Plotting test accuracies for each model
plt.figure(figsize=(10, 5))
plt.bar(test_accuracies.keys(), test_accuracies.values(), color='skyblue')
plt.xlabel('Model')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy of Different Models')
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
